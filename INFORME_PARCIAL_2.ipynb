## **1. AVL vs Arboles Binario**

### **Descripci√≥n del funcionamiento de los √Årboles AVL y su diferencia con los √Årboles Binarios de B√∫squeda.**

Un √°rbol binario de b√∫squeda (BST) es una estructura donde cada nodo tiene dos hijos (izquierdo y derecho), y est√° organizado de tal manera que:

Todos los nodos en el sub√°rbol izquierdo son menores que el nodo principal.
Todos los nodos en el sub√°rbol derecho son mayores.
El problema es que, aunque normalmente permite b√∫squedas r√°pidas, si el √°rbol se desbalancea (uno de los lados crece mucho m√°s que el otro), las b√∫squedas se vuelven lentas. En los peores casos, el rendimiento de las b√∫squedas y otras operaciones puede llegar a ser igual que en una lista enlazada.

Por otro lado, los √°rboles AVL son √°rboles binarios de b√∫squeda que se mantienen balanceados. Esto significa que la diferencia de altura entre los sub√°rboles de cualquier nodo es siempre como m√°ximo 1. Para lograr esto, los √°rboles AVL realizan rotaciones autom√°ticas despu√©s de insertar o eliminar nodos, manteniendo el √°rbol siempre equilibrado.

### **Diferencias principales:**



#### **Balanceo:**

- **BST:** No garantizan un balanceo, por lo que pueden desbalancearse si las inserciones siguen un patr√≥n desfavorable, como insertar nodos en orden ascendente o descendente. Esto puede provocar que el √°rbol se asemeje a una lista enlazada, lo que afecta negativamente el rendimiento.

- **AVL:** Se autoajustan para mantenerse balanceados. Despu√©s de cada inserci√≥n o eliminaci√≥n, el √°rbol comprueba el equilibrio y, si es necesario, realiza rotaciones para asegurar que la diferencia de altura entre los sub√°rboles izquierdo y derecho de cualquier nodo nunca sea mayor que 1.

#### **Rendimiento:**

- **BST:** En su forma ideal, ofrecen b√∫squedas, inserciones y eliminaciones en tiempo O(log n), pero en el peor de los casos, si se desbalancean, las operaciones pueden degradarse a O(n), lo que significa que pierden eficiencia.

- **AVL:** Gracias a su balanceo constante, las operaciones se realizan siempre en O(log n), lo que garantiza un rendimiento √≥ptimo incluso en el peor de los casos, independientemente del orden de inserci√≥n o eliminaci√≥n.

#### **Rotaciones:**

- **BST:** No realizan rotaciones, lo que permite que el √°rbol se desbalancee sin intervenci√≥n.

- **AVL:** Utilizan rotaciones simples (izquierda o derecha) y dobles (izquierda-derecha o derecha-izquierda) para restaurar el equilibrio del √°rbol. Estas rotaciones son r√°pidas y se aplican localmente en las √°reas desbalanceadas.

##### **Altura del √°rbol:**

- **BST:** La altura puede crecer de manera desproporcionada en √°rboles desbalanceados, aumentando la profundidad y por lo tanto el tiempo necesario para realizar b√∫squedas o inserciones.

- **AVL:** La altura se mantiene cerca de log(n) debido al balanceo, lo que asegura que las operaciones de b√∫squeda, inserci√≥n y eliminaci√≥n se mantengan eficientes.

#### **Costo de mantenimiento:**

- **BST:** No tienen un costo adicional por mantener el equilibrio, ya que no realizan rotaciones. Esto los hace m√°s simples en su implementaci√≥n, pero menos eficientes cuando se desbalancean.
- **AVL:** Tienen un peque√±o costo adicional debido a las rotaciones despu√©s de cada inserci√≥n o eliminaci√≥n, pero este costo es m√≠nimo y asegura que el rendimiento general del √°rbol sea √≥ptimo a largo plazo.

## **2. Rotaciones**


### **Implementacion de rotaciones para mantener el balance en un AVL**

En un √°rbol AVL, las **rotaciones** se implementan para restaurar el balance despu√©s de una inserci√≥n o eliminaci√≥n que cause un desbalance en el √°rbol. El balance de un nodo se determina por la diferencia de alturas entre su sub√°rbol izquierdo y derecho (balance factor), y debe estar entre -1 y 1. Si el balance excede estos l√≠mites, se ejecuta una rotaci√≥n.

Las rotaciones b√°sicas son:

#### **Rotacion Simple Derecha (LL):**
*  Ocurre cuando un nodo tiene un sub√°rbol izquierdo m√°s pesado.
*  El sub√°rbol derecho del hijo izquierdo pasa a ser el sub√°rbol izquierdo del nodo desbalanceado, y el hijo izquierdo toma su lugar.

![alt text](https://player.slideplayer.es/8/2438423/data/images/img4.jpg)

#### **Rotacion Simple Izquierda (RR):**
*  Se utiliza cuando un nodo tiene un sub√°rbol derecho m√°s pesado.
*  Se mueve el sub√°rbol izquierdo del hijo derecho como sub√°rbol derecho del
nodo desbalanceado, y el hijo derecho toma la posici√≥n del nodo desbalanceado.

![alt text](https://player.slideplayer.es/8/2438423/data/images/img2.png)

#### **Rotacion Izquierda Derecha (LR):**
* Cuando un sub√°rbol izquierdo del nodo est√° desbalanceado hacia la derecha.
* Se realiza una rotaci√≥n simple a la izquierda sobre el hijo izquierdo, seguida de
una rotaci√≥n simple a la derecha sobre el nodo desbalanceado.

![alt text](https://player.slideplayer.es/8/2438423/data/images/img6.png)

#### **Rotacion Derecha Izquierda (RL):**
* Aparece cuando el sub√°rbol derecho est√° desbalanceado hacia la izquierda.
* Se realiza una rotaci√≥n a la derecha sobre el hijo derecho, luego una rotaci√≥n a
la izquierda sobre el nodo desbalanceado

        balance = self._obtener_balance(nodo_actual.padre)

        if abs(balance) > 1:
            camino = [nodo_actual.padre] + camino
            self._reequilibrar_nodo(camino[0], camino[1], camino[2])

````python
import sys

# Clase Nodo que representa cada nodo en el √°rbol
class Nodo:
    def __init__(self, valor, padre=None):
        self.valor = valor
        self.padre = padre
        self.izquierda = None
        self.derecha = None
        self.altura = 1

# Clase que representa un √Årbol AVL
class ArbolAVL:
    def __init__(self):
        self.raiz = None

    # M√©todo para insertar un nodo en el √°rbol
    def insertar(self, valor, nodo_actual):
        if not self.raiz:
            self.raiz = Nodo(valor)
            return

        if valor < nodo_actual.valor:
            if nodo_actual.izquierda:
                self.insertar(valor, nodo_actual.izquierda)
            else:
                nodo_actual.izquierda = Nodo(valor, nodo_actual)
                self._revisar_insertar(nodo_actual.izquierda)
        elif valor > nodo_actual.valor:
            if nodo_actual.derecha:
                self.insertar(valor, nodo_actual.derecha)
            else:
                nodo_actual.derecha = Nodo(valor, nodo_actual)
                self._revisar_insertar(nodo_actual.derecha)
        else:
            print("El Valor que estas insertando ya se encuentra en el arbol")


    # M√©todo para verificar e inspeccionar si se necesita reequilibrar tras una inserci√≥n
    def _revisar_insertar(self, nodo_actual, camino=[]):
        if nodo_actual.padre is None: return
        camino = [nodo_actual] + camino

        balance = self._obtener_balance(nodo_actual.padre)

        if abs(balance) > 1:
            camino = [nodo_actual.padre] + camino
            self._reequilibrar_nodo(camino[0], camino[1], camino[2])
            return

        nueva_altura = 1 + nodo_actual.altura
        if nueva_altura > nodo_actual.padre.altura:
            nodo_actual.padre.altura = nueva_altura

        self._revisar_insertar(nodo_actual.padre, camino)


    # M√©todo para obtener el balance de un nodo
    def _obtener_balance(self, nodo_actual):
        if nodo_actual is None:
            return 0
        return self._obtener_altura(nodo_actual.izquierda) - self._obtener_altura(nodo_actual.derecha)

    # M√©todo para obtener la altura de un nodo
    def _obtener_altura(self, nodo_actual):
        if nodo_actual is None: return 0
        return nodo_actual.altura


    # M√©todo para reequilibrar el √°rbol en caso de que se desequilibre
    def _reequilibrar_nodo(self, z, y, x):
        if y == z.izquierda and x == y.izquierda:
            self._rotar_derecha(z)
        elif y == z.izquierda and x == y.derecha:
            self._rotar_izquierda(y)
            self._rotar_derecha(z)
        elif y == z.derecha and x == y.derecha:
            self._rotar_izquierda(z)
        elif y == z.derecha and x == y.izquierda:
            self._rotar_derecha(y)
            self._rotar_izquierda(z)
        else:
            raise Exception('_reequilibrar_nodo: ¬°Configuraci√≥n de nodos z, y, x no reconocida!')


    # Rotaci√≥n a la derecha
    def _rotar_derecha(self, y):
        sub_raiz = y.padre
        z = y.izquierda
        t3 = z.derecha
        z.derecha = y
        y.padre = z
        y.izquierda = t3

        if t3 is not None:
            t3.padre = y

        z.padre = sub_raiz

        if z.padre is None:
            self.raiz = z
        else:
            if z.padre.izquierda == y:
                z.padre.izquierda = z
            else:
                z.padre.derecha = z

        y.altura = 1 + max(self._obtener_altura(y.izquierda), self._obtener_altura(y.derecha))
        z.altura = 1 + max(self._obtener_altura(z.izquierda), self._obtener_altura(z.derecha))


    # Rotaci√≥n a la izquierda
    def _rotar_izquierda(self, z):
        sub_raiz = z.padre
        y = z.derecha
        t2 = y.izquierda
        y.izquierda = z
        z.padre = y
        z.derecha = t2

        if t2 is not None:
            t2.padre = z

        y.padre = sub_raiz

        if y.padre is None:
            self.raiz = y
        else:
            if y.padre.izquierda == z:
                y.padre.izquierda = y
            else:
                y.padre.derecha = y

        z.altura = 1 + max(self._obtener_altura(z.izquierda), self._obtener_altura(z.derecha))
        y.altura = 1 + max(self._obtener_altura(y.izquierda), self._obtener_altura(y.derecha))


    # M√©todo para eliminar un nodo del √°rbol
    def eliminar(self, valor, nodo_actual):
        if not self.raiz:
            print("El √°rbol est√° vac√≠o, no se puede eliminar.")
            return

        if nodo_actual is None:
            print("El nodo no existe en el √°rbol.")
            return nodo_actual

        if valor < nodo_actual.valor:
            nodo_actual.izquierda = self.eliminar(valor, nodo_actual.izquierda)
        elif valor > nodo_actual.valor:
            nodo_actual.derecha = self.eliminar(valor, nodo_actual.derecha)
        else:
            if nodo_actual.izquierda is None:
                temp = nodo_actual.derecha
                nodo_actual = None
                return temp
            elif nodo_actual.derecha is None:
                temp = nodo_actual.izquierda
                nodo_actual = None
                return temp

            temp = self.obtener_minimo(nodo_actual.derecha)
            nodo_actual.valor = temp.valor
            nodo_actual.derecha = self.eliminar(temp.valor, nodo_actual.derecha)

        if nodo_actual is None:
            return nodo_actual

        nodo_actual.altura = 1 + max(self._obtener_altura(nodo_actual.izquierda), self._obtener_altura(nodo_actual.derecha))

        balance = self._obtener_balance(nodo_actual)

        if balance > 1 and self._obtener_balance(nodo_actual.izquierda) >= 0:
            return self._rotar_derecha(nodo_actual)

        if balance > 1 and self._obtener_balance(nodo_actual.izquierda) < 0:
            nodo_actual.izquierda = self._rotar_izquierda(nodo_actual.izquierda)
            return self._rotar_derecha(nodo_actual)

        if balance < -1 and self._obtener_balance(nodo_actual.derecha) <= 0:
            return self._rotar_izquierda(nodo_actual)

        if balance < -1 and self._obtener_balance(nodo_actual.derecha) > 0:
            nodo_actual.derecha = self._rotar_derecha(nodo_actual.derecha)
            return self._rotar_izquierda(nodo_actual)

        return nodo_actual

    # M√©todo para obtener el valor m√≠nimo en el √°rbol
    def obtener_minimo(self, nodo_actual):
        if nodo_actual is None or nodo_actual.izquierda is None:
            return nodo_actual
        return self.obtener_minimo(nodo_actual.izquierda)


    # M√©todo para obtener el valor m√°ximo en el √°rbol
    def obtener_maximo(self, nodo_actual):
        if nodo_actual is None or nodo_actual.derecha is None:
            return nodo_actual
        return self.obtener_maximo(nodo_actual.derecha)


    # M√©todo para obtener la ra√≠z del √°rbol
    def obtener_raiz(self):
        return self.raiz.valor

    # Recorrido en orden del √°rbol
    def recorrido_in_order(self, nodo):
        if nodo:
            self.recorrido_in_order(nodo.izquierda)
            print(nodo.valor, end=" - ")
            self.recorrido_in_order(nodo.derecha)


    # Recorrido en preorden del √°rbol
    def recorrido_pre_order(self, nodo):
        if nodo:
            print(nodo.valor, end=" - ")
            self.recorrido_pre_order(nodo.izquierda)
            self.recorrido_pre_order(nodo.derecha)


    # Recorrido en postorden del √°rbol
    def recorrido_post_order(self, nodo):
        if nodo:
            self.recorrido_post_order(nodo.izquierda)
            self.recorrido_post_order(nodo.derecha)
            print(nodo.valor, end=" - ")

    # Metodo para buscar en el arbol
    def buscar(self, nodo, valor):
        if not nodo or nodo.valor == valor:
            return nodo
        if valor < nodo.valor:
            return self.buscar(nodo.izquierda, valor)
        return self.buscar(nodo.derecha, valor)


    # M√©todo para imprimir el √°rbol
    def imprimir_arbol(self, nodo, nivel=0):
        if nodo is not None:
            self.imprimir_arbol(nodo.derecha, nivel + 1)
            print(" " * 4 * nivel + "-> " + str(nodo.valor))
            self.imprimir_arbol(nodo.izquierda, nivel + 1)


def main():
    arbol = ArbolAVL()

    while True:

        print("\n"
            +"-" * 40
            + "\n\n\t Men√∫ del √Årbol AVL\n"
            + "\n1. ¬øEsta vacio el Arbol?"
            + "\n2. Insertar un nodo"
            + "\n3. Eliminar un nodo"
            + "\n4. Buscar un nodo"
            + "\n5. Recorrido en preorden"
            + "\n6. Recorrido en inorden"
            + "\n7. Recorrido en postorden"
            + "\n8. Obtener Raiz"
            + "\n9. Obtener Valor Maximo del Arbol"
            + "\n10. Obtener Valor Minimo del Arbol"
            + "\n11. Obtener Factor de Balance de Cierto Nodo"
            + "\n12. Obtener Altura Total del Arbol"
            + "\n13. Imprimir √°rbol"
            + "\n0. Salir\n\n"
            + "-" * 40)

        opcion = input("\nSelecciona una opci√≥n: ")

        if opcion == "1":
            if arbol.raiz is None:
                print("\nEl √°rbol est√° vac√≠o.")
            else:
                print("\nEl √°rbol no est√° vac√≠o.")

        elif opcion == "2":
            nodo = int(input("\nIngrese el valor del nodo: "))
            arbol.insertar(nodo, arbol.raiz)
            print(f"Nodo {nodo} insertado.")

        elif opcion == "3":
            nodo = int(input("\nIngrese el valor del nodo a eliminar: "))
            arbol.eliminar(nodo, arbol.raiz)
            print(f"Nodo {nodo} eliminado.")

        elif opcion == "4":
            nodo = int(input("\nIngrese el nodo a buscar: "))
            resultado = arbol.buscar(arbol.raiz, nodo)
            print(f"B√∫squeda del nodo {nodo}: {'Encontrado' if resultado else 'No encontrado'}")

        elif opcion == "5":
            print("\nRecorrido en preorden: ", end="")
            arbol.recorrido_pre_order(arbol.raiz)
            print()

        elif opcion == "6":
            print("\nRecorrido en inorden: ", end="")
            arbol.recorrido_in_order(arbol.raiz)
            print()

        elif opcion == "7":
            print("\nRecorrido en postorden: ", end="")
            arbol.recorrido_post_order(arbol.raiz)
            print()

        elif opcion == "8":
            if arbol.raiz is None:
                print("\nEl √°rbol est√° vac√≠o.")
            else:
                print("\nRa√≠z del √°rbol: ", arbol.obtener_raiz())

        elif opcion == "9":
            print(f"\nValor Maximo del Arbol: {arbol.obtener_maximo(arbol.raiz).valor}")

        elif opcion == "10":
            print(f"\nValor Minimo del Arbol: {arbol.obtener_minimo(arbol.raiz).valor}")

        elif opcion == "11":
            nodo = int(input("\nIngrese el nodo para obtener el factor de balance: "))
            print(f"\nEl Factor de balance del nodo {nodo} es {arbol._obtener_balance(arbol.buscar(arbol.raiz, nodo))}")

        elif opcion == "12":
            print(f"\nLa Altura Total del Arbol es {arbol._obtener_altura(arbol.raiz)}")

        elif opcion == "13":
            print("\n√Årbol AVL actual:\n")
            arbol.imprimir_arbol(arbol.raiz)

        elif opcion == "0":
            print("Saliendo del programa.")
            break

        else:
            print("Opci√≥n inv√°lida. Int√©ntalo de nuevo.")

# Funci√≥n principal para ejecutar el programa
if __name__ == "__main__":
    main()
````
